# 7. Verificando os dados por estado

```python
import re

fig, ax = plt.subplots(figsize=(16, 8))
plt.suptitle('Nota Redação x Escolaridade x Estado', size= 20)

# Filtrar e processar os dados
dados_filtrados = dados.filter(items=['Nota da prova de redação.',
                                      'Até que série sua mãe, ou a mulher responsável por você, estudou?',
                                      'Sigla da Unidade da Federação da escola']) \
                        .where(dados['Até que série sua mãe, ou a mulher responsável por você, estudou?'] != 'Nunca estudou.') \
                        .groupby(['Até que série sua mãe, ou a mulher responsável por você, estudou?',
                                  'Sigla da Unidade da Federação da escola']) \
                        .mean().sort_values(by='Nota da prova de redação.', ascending=False) \
                        .unstack()

# Plotar o gráfico com rótulos rotacionados e limitando as categorias exibidas
dados_filtrados.plot(ax=ax)
plt.xticks(rotation=45)
plt.xlim(0, len(dados_filtrados) - 1)  # Defina o limite do eixo x para o número de categorias - 1



# Edita Legenda para aparecer aapenas o UF dos estados
handles, labels = ax.get_legend_handles_labels()
edited_labels = [re.search(',\s(.+?)\)', label).group(1) for label in labels] #Expressão Regular para pegar apenas a UF
ax.legend(edited_labels, bbox_to_anchor=(1.1, 1), loc=0, borderaxespad=0.)


# Exibir o gráfico
ax.grid(color='gray')
plt.show()
```
